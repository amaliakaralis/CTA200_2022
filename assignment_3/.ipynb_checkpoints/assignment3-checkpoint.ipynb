## Question 1

```python
import numpy as np
import matplotlib.pyplot as plt
from complex_iter import complex_iter
```

```python
N_pts = 10
N_iter = 15
z_0 = 0
# get the XY grid, Z values and div_iter values for an N_ptsxN_pts grid in the complex plane
# and N_iter iterations of z
X, Y, Z, div_iter = complex_iter(z_0, N_pts, N_iter)
# first, let's look at which points in the complex plane diverge after N_iter iterations
# in this case, we only care about the last point in each z iteration
Z1 = Z[:, -1]
```

```python
# plot the complex plane, differentiating between points that diverge in z and points that do not
plt.figure(figsize=(6, 5))
plt.plot(X[np.isnan(Z1)], Y[np.isnan(Z1)], 'r^', label = 'Diverging')
plt.plot(X[np.isfinite(Z1)], Y[np.isfinite(Z1)], 'ko', label = 'Finite')
#plt.legend(bbox_to_anchor=(0., -0.2), loc='center left', fontsize=15)
plt.xlabel('X', fontsize=20)
plt.ylabel('Y', fontsize=20)
plt.title('After %i iterations' % N_iter, fontsize=20)
plt.savefig('Q1_Fig1.pdf')
plt.show()
```

```python
# now, color the diverging points based on the iterations at which they diverge

plt.figure(figsize=(7.4,5))
plt.scatter(X[np.isnan(Z1)], Y[np.isnan(Z1)], marker='^', c = div_iter[np.isfinite(div_iter)], label = 'Diverging',
           cmap = 'plasma')
plt.plot(X[np.isfinite(Z1)], Y[np.isfinite(Z1)], 'ko', label = 'Finite')
clb = plt.colorbar()
clb.set_label(label='Iteration of Divergence', fontsize=15)
plt.xlabel('X', fontsize=20)
plt.ylabel('Y', fontsize=20)
plt.title('After %i Iterations' % N_iter, fontsize=20)
#plt.legend(bbox_to_anchor=(0., -0.3), loc='center left', fontsize=15)
plt.savefig('Q1_Fig2.pdf')
plt.show()
```

## Question 2

```python
# the convection equations (Eqns. 25, 26, 27 of Lorentz (1963))
def W_dot(t, W, params):
    """
    The three coupled ordinary differential equations that represent the convection equations.
    
    Parameters:
    -----------
    t: numpy array
        dimensionless time 
    W: [X, Y, Z]
        X: proportional to the intensity of the convective motion
        Y: proportional to the temperature difference between the ascending and descending currents
        Z: proportional to the distortion of the vertical temperature profile from linearity
    params: [sigma, r, b]
        sigma: Prandtl number 
        r: Rayleigh number
        b: dimensionless lenght scale
    
    Returns:
    --------
    W_dot: [X_dot, Y_dot, Z_dot] as defined in Eqns 25, 26, 27 in Lorentz (1963)
    """
    sigma, r, b = params
    return [-sigma*(W[0]-W[1]), r*W[0] - W[1] - W[0]*W[2], -b*W[2] + W[0]*W[1]]
```

```python
from scipy.integrate import solve_ivp
N = 1000
dt = 0.01
t_eval = np.linspace(0, 30, N)
sol = solve_ivp(W_dot, (0, 60), [0., 1., 0.], args = [[10., 28., 8./3.]], t_eval = t_eval)
tau = sol.t # the dimensionless time parameter
N = tau/dt # Lorentz uses N to label his plots
X = sol.y[0]
Y = sol.y[1]
Z = sol.y[2]
```

```python
# reproduce Figure 1 from Lorentz (1963)
fig, ax = plt.subplots(3, 1, figsize=(15, 9))

ax[0].plot(N[:333], Y[:333])
ax[0].axhline(y=0, color='black')
ax[0].set_ylabel('Y', fontsize=20)

ax[1].plot(N[333:666], Y[333:666])
ax[1].axhline(y=0, color='black')
ax[1].set_ylabel('Y', fontsize=20)

ax[2].plot(N[666:], Y[666:])
ax[2].axhline(y=0, color='black')
ax[2].set_ylabel('Y', fontsize=20)
ax[2].set_xlabel(r'N = t/$\Delta$ t', fontsize=20)

plt.savefig('Lorentz_fig1.pdf')
plt.show()
```

```python
Npts = 1001
dt = 0.01
t_eval2 = np.linspace(14, 19, Npts)
sol2 = solve_ivp(W_dot, (0, 60), [0., 1., 0.], args = [[10., 28., 8./3.]], t_eval = t_eval2)
tau2 = sol2.t # the dimensionless time parameter
N2 = tau2/dt # Lorentz uses N to label his plots
X2 = sol2.y[0]
Y2 = sol2.y[1]
Z2 = sol2.y[2]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
annotations = ['14', '15', '16', '17', '18', '19']

ax[0].plot(Y2, Z2)
ax[0].set_ylabel('Z', fontsize=20)
ax[0].set_xlabel('Y', fontsize=20)
ax[0].scatter(Y2[tau2%1==0], Z2[tau2%1==0])
for i, label in enumerate(annotations):
    ax[0].annotate(label, (Y2[tau2%1==0][i], Z2[tau2%1==0][i]))


ax[1].plot(Y2, X2)
ax[1].scatter(Y2[tau2%1==0], X2[tau2%1==0])
for i, label in enumerate(annotations):
    ax[1].annotate(label, (Y2[tau2%1==0][i], X2[tau2%1==0][i]))
    
ax[1].set_ylim(ax[1].get_ylim()[::-1])
ax[1].set_ylabel('X', fontsize=20)
ax[1].set_xlabel('Y', fontsize=20)

plt.savefig('Lorentz_fig2.pdf')
plt.show()
```

```python
Npts = 1000
dt = 0.01
t_eval = np.linspace(0, 30, Npts)
sol_prime = solve_ivp(W_dot, (0, 60), [0., 1.+1e-8, 0.], args = [[10., 28., 8./3.]], t_eval = t_eval)
tau_prime = sol_prime.t # the dimensionless time parameter
N_prime = tau/dt # Lorentz uses N to label his plots
X_prime = sol_prime.y[0]
Y_prime = sol_prime.y[1]
Z_prime = sol_prime.y[2]
```

```python
fig, ax = plt.subplots(3, 1, figsize=(15, 9))
ax[0].plot(tau, abs(X_prime-X))
ax[0].set_ylabel("abs(X' - X)", fontsize=20)
ax[0].set_yscale('log')

ax[1].plot(tau, abs(Y_prime-Y))
ax[1].set_ylabel("abs(Y' - Y)", fontsize=20)
ax[1].set_yscale('log')

ax[2].plot(tau, abs(Z_prime-Z))
ax[2].set_ylabel("abs(Z' - Z)", fontsize=20)
ax[2].set_xlabel('Time', fontsize=20)
ax[2].set_yscale('log')

plt.savefig('Error_plot.pdf')
plt.show()
```
